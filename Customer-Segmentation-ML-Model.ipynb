```python
X = data[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
X
```

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans()
k_means.fit(X)
```

```python
k_means = KMeans(n_clusters=5)
k_means.fit_predict(X)
```

```python
wcss = []
for i in range(1,11):
    k_means = KMeans(n_clusters=i)
    k_means.fit(X)
    wcss.append(k_means.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of Customers")
plt.ylabel("WCSS")
plt.show()
plt.savefig("Elbow Method.png")
```

```python
X = data[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
k_means = KMeans(n_clusters=5,random_state=42)
y_means = k_means.fit_predict(X)
```

```python
y_means
```

```python
plt.scatter(X.iloc[y_means==0,0],X.iloc[y_means==0,1],s=100,c='red',label='Budget Keepers')
plt.scatter(X.iloc[y_means==1,0],X.iloc[y_means==1,1],s=100,c='orange',label='Occasional Spenders')
plt.scatter(X.iloc[y_means==2,0],X.iloc[y_means==2,1],s=100,c='yellow',label='Steady Shoppers')
plt.scatter(X.iloc[y_means==3,0],X.iloc[y_means==3,1],s=100,c='blue',label='Value Maximizers')
plt.scatter(X.iloc[y_means==4,0],X.iloc[y_means==4,1],s=100,c='green',label='Premium Elites')
plt.scatter(k_means.cluster_centers_[:,0],k_means.cluster_centers_[:,1],s=100,c='black')
plt.title("Customer Segementation")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```

```python
plt.savefig("Customer-Segmented-Chart.png")
```
